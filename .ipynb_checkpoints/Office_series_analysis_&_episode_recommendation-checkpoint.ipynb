```python
import pandas as pd
import numpy as np
import six
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import pylab as pl
import re
import random
import nltk
from sklearn.metrics.pairwise import cosine_similarity
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
import warnings
warnings.filterwarnings('ignore')
```

```python
df_lines = pd.read_csv("TheOfficeLinesPerCharacter.csv")
df_rat = pd.read_csv("TheOfficeIMDBPerEpisode.csv")
```

```python
df_lines.head()
```

```python
df_rat.head()
```

```python
df_main = pd.DataFrame(columns = ["Season", "Episode", "Rating","Votes","WrittenBy"])
characters = df_lines.speaker.value_counts().reset_index(name="count").query("count > 100")["index"]

episode = 1
season = 1
for i in range(len(df_rat.index)):
    if(season != df_rat.iloc[i]['Season']):
        episode = 1
        season += 1 
    df_main = df_main.append({'Season': season, 'Episode': episode, 'Rating':df_rat.iloc[i]['Rating'], 'Votes':df_rat.iloc[i]['Num_Votes'],'WrittenBy' : df_rat.iloc[i]['WrittenBy'],}, ignore_index=True)
    episode += 1
      
#Add characters columns
for character in characters:
    df_main[str(character)] = np.NaN
    
#Add lines spoken count for each character
for j in range(len(df_lines.index)):
    speaker = df_lines.iloc[j]['speaker']
    if speaker in df_main.columns:
        season = df_lines.iloc[j]['season']
        episode = df_lines.iloc[j]['episode']
        index = df_main.loc[df_main['Season']==season].loc[df_main['Episode']==episode].index.values.astype(int)[0]
        if(pd.isna(df_main.at[index,speaker])):
            df_main.at[index,speaker] = 1
        else:
            df_main.at[index,speaker] += 1
```

```python
TF = (df_main.isnull())

for col in TF.columns:
        if (TF[col].value_counts()[0] < 10):
            df_main = df_main.drop(columns=[col])
df_main.to_csv("./main.csv")
df_main.head()
```

```python
df_main = pd.read_csv("main.csv")
hist = df_main.hist(column = "Rating", sharex=True, sharey=True, figsize=(10,5), grid=False)
pl.title("Histogram of The Office Episode Ratings")
pl.ylabel("Frequency")
pl.xlabel("Episode Rating")
```

```python
df_condensed = pd.DataFrame(columns = ["character", "total lines", "avg lines per appearance"])
index_no = df_main.columns.get_loc('Michael')
index_no
total_line = df_main.sum(axis=0)
for k in range(6, df_main.shape[1]):
    character = df_main.columns[k]
    total_appearence = len(df_main[character].loc[df_main[character] != 0])
    
    row = {"character" : character, "total lines" : int(total_line[k]), "avg lines per appearance" : int(total_line[k]/total_appearence)}
    df_condensed = df_condensed.append(row, ignore_index=True)
```

```python
cor = []

for char in df_condensed['character']:
    character = char
    cor.append( round(df_main['Rating'].corr(df_main[char]),4) )
    
    df1 = df_main[['Rating',character]].dropna()
    x = np.array(df1[character])
    y = np.array(df1['Rating'])
    dem = x.dot(x) - x.mean() * x.sum()
    m = (x.dot(y) - y.mean() * x.sum())/dem
    b = (y.mean() * x.dot(x) - x.mean() * x.dot(y))/dem
    line = m*x + b
    df_main.plot(x = character, y='Rating', kind = 'scatter', color='#ff9999', legend=None)
    plt.plot(x, line, '#96c7d9')
    plt.xlabel(character + "'s lines spoken", fontsize=12)
    plt.ylabel("IMDB episode rating")
    plt.show()
```

```python
data = pd.read_csv('the_office_series.csv')
data.head()
```

```python
stop_words = stopwords.words('english')
stemmer = PorterStemmer()
```

```python
summary = []
for i in range(0,len(data)):
    about = re.sub('[^0-9a-zA-Z]', ' ', data['About'][i])
#     about = re.sub(r'[\W_]+', '',data['About'][i] )
    about = about.lower()
    about = about.split()
    about = [stemmer.stem(word) for word in about if not word in stop_words]
    about = ' '.join(about)
    summary.append(about)
    dirs = data['Director']
    writs =data['Writers']
    
# Adding Directors and Writers


Description = []
for k in range(0,len(data)):
    Description.append(summary[k]+dirs[k]+writs[k])
```

```python
tf = TfidfVectorizer()
X = tf.fit_transform(Description)

cos_sim = cosine_similarity(X, X)
```

```python
cos_sim_df = pd.DataFrame(data=cos_sim,columns=data.EpisodeTitle)
cos_sim_df.head()
```

```python
def get_recommendation(episode_name):
    for i in range(0,len(data)):
        
        if data.values[i][2] == episode_name:
            episode_index = data.values[i][0]
            sorted_data = cos_sim_df.iloc[episode_index].sort_values(ascending=False)
            sorted_df = pd.DataFrame(sorted_data)
            print(sorted_df.iloc[1:6])
```

```python
get_recommendation('Michael Scott Paper Company')
```
